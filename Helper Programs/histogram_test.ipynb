```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in image
img = cv2.imread("lsun/test/0.jpg")

# Display image
plt.imshow(img)
plt.show()
```

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = [x.flatten() for x in cv2.split(img)]
H, W = lab.shape[:2]
```

```python
bins = 25
plt.hist(l,bins,[0,256])
plt.title("L Channel")
plt.show()
prob = l.astype("float32") / (H * W)
clipped = prob.clip(np.finfo(float).eps, 256-np.finfo(float).eps)
entropy = -sum(clipped * np.log(clipped))
print("Entropy:", entropy)

plt.hist(a,bins,[0,256])
plt.title("A Channel")
plt.show()
prob = a.astype("float32") / (H * W)
clipped = prob.clip(np.finfo(float).eps, 256-np.finfo(float).eps)
entropy = -sum(clipped * np.log(clipped))
print("Entropy:", entropy)

plt.hist(b,bins,[0,256])
plt.title("B Channel")
plt.show()
prob = b.astype("float32") / (H * W)
clipped = prob.clip(np.finfo(float).eps, 256-np.finfo(float).eps)
entropy = -sum(clipped * np.log(clipped))
print("Entropy:", entropy)
```
