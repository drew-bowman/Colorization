```python
import os
import cv2
import random
```

```python
def list_image_files(directory):
    files = sorted(os.listdir(directory))
    return [os.path.join(directory, f) for f in files if is_an_image_file(f)]

def is_an_image_file(filename):
    IMAGE_EXTENSIONS = ['.png', '.jpg', '.jpeg']
    for ext in IMAGE_EXTENSIONS:
        if ext in filename:
            return True
    return False
```

```python
folder_names = ['badlands/', 'butte/', 'canyon/', 'cliff/', 'field-wild/', 'forest-broadleaf/', 'forest_path/', 'lake-natural/', 'mountain/', 'tundra/']
```

```python
starting_dir = "../places365_standard/train/"
for i in folder_names:
    print("Structuring images in ", i, "...", sep="")
    
    img_paths = list_image_files(starting_dir + i)
    print("Before shuffling...")
    for img in img_paths[:3]:
        print(img)

    random.shuffle(img_paths)

    print("\nAfter shuffling...")
    for img in img_paths[:3]:
        print(img)
    
    # Take first 100 images for testing
    count = 0
    for img_path in img_paths[:100]:
        save_path = "../places2/test/" + i[:-1] + "_" + str(count) + ".jpg"

        img = cv2.imread(img_path)
        cv2.imwrite(save_path, img)

        count += 1

    print("All testing images saved!\n")

    count = 0
    # Ues remaining images for training
    for img_path in img_paths[100:]:
        save_path = "../places2/train/" + i[:-1] + "_" + str(count) + ".jpg"

        img = cv2.imread(img_path)
        cv2.imwrite(save_path, img)

        count += 1

        if count+1 % 1000 == 0:
            print(count+1, "training images saved")
    
    print(count+1, "training images saved\n")
        
print("All training images saved!")
```

```python
test_images = list_image_files("../places2/test/")
print("There are", len(test_images), "test images")

train_images = list_image_files("../places2/train/")
print("There are", len(train_images), "training images")
```
